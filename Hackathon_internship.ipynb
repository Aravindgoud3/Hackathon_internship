```python
data['input']=data['input']**2
data
```

```python
fv=data[['input']]
cl=data['output']
x_train,x_test,y_train,y_test=train_test_split(fv,cl,test_size=0.1,random_state=0)
std=StandardScaler()
std_xtrain=std.fit_transform(x_train)
std_xtest=std.transform(x_test)
model=LinearRegression()
model.fit(std_xtrain,y_train)
predi=model.predict(std_xtest)
```

```python
pd.DataFrame({'Actual':y_test,'Predicted':predi})
```

```python
print('mean_squared_error:',mean_squared_error(y_test,predi))
print('mean_absolute_error:',mean_absolute_error(y_test,predi))
print('r2_score:',r2_score(y_test,predi))
```
